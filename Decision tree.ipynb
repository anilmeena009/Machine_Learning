# Decision Tree
A decsion tree is a supervised machine learning model that is used for classification and regression task,thats why it is considerd as CART.
It was like a flow chart or tree structure where each internal mode represent a feature and
each branch represent a decsion or condition and each leef node represent and output or class label in the classificationor a continues value in the regression.
A decsion tree split the data intop subsets based on the value of input features and this process is repeted recursly forming a tree line structure

# How it works
It start with the entire data it chose the best feature to split the data based on a matrix.
like --gini-index,entropy,mse(mean squre erroe)
Then it will split the data set into smaller subset then it will repet the process for each child node until.A stoping condition is mat.


```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("covid_toy.csv")
df.head(2)
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender'] = lb. fit_transform(df['gender' ])
df['cough'] = lb. fit_transform(df['cough' ])
df['city'] = lb. fit_transform(df['city' ])
df['has_covid'] = lb. fit_transform(df['has_covid' ])
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn. tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train , y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
